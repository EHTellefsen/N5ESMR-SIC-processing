```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
FIRSTDATE = '1972-12'
LASTDATE = '1977-05'
monthRange = np.arange(np.datetime64(FIRSTDATE,'M'),np.datetime64(LASTDATE,'M'))
SIE_NH_30_LDTP =  pd.read_csv('N5ESMR_SIE_LDTP_NH_197212-197705.csv')
SIE_NH_30_Kolbe = pd.read_csv('N5ESMR_SIE_Kolbe_NH_197212-197705.csv')
SIE_SH_30_LDTP =  pd.read_csv('N5ESMR_SIE_LDTP_SH_197212-197705.csv')
SIE_SH_30_Kolbe = pd.read_csv('N5ESMR_SIE_Kolbe_SH_197212-197705.csv')
```

```python
fig, ax = plt.subplots(2,1,figsize=(13,7),sharex=True,layout='tight')
ax[0].plot(monthRange,SIE_NH_30_Kolbe['SIE_NH_30%[km^2]']/1e6,'r-',linewidth=2, label='Kolbe et al.')
ax[0].plot(monthRange,SIE_NH_30_LDTP['SIE_NH_30%[km^2]']/1e6,'b:',linewidth=2, label='LDTP')
ax[0].grid()
ax[0].set_title('Arctic sea ice extent 1972-1977')
#ax[0].set_ylabel('Sea ice extend, SIC>30% [mill. $km^2$]')
#ax[0].set_xlabel('Date')
ax[0].set_ylim([0,20])
ax[0].legend(loc='lower right', fontsize=10)

ax[1].plot(monthRange,SIE_SH_30_Kolbe['SIE_SH_30%[km^2]']/1e6,'r-',linewidth=2)
ax[1].plot(monthRange,SIE_SH_30_LDTP['SIE_SH_30%[km^2]']/1e6,'b:',linewidth=2)
ax[1].grid()
ax[1].set_title('Antarctic sea ice extent 1972-1977')
#ax[1].set_ylabel('Sea ice extend, SIC>30% [mill. $km^2$]')
ax[1].set_xlabel('Month')
ax[1].set_ylim([0,20])

fig.supylabel('Sea ice extent (SIC>30%) [mill. $km^2$]')
plt.savefig('N5ESMR_SIE.pdf', bbox_inches='tight')
```
